## Logistic regression on the titanic dataset

```python
titanic = pd.read_csv("titanic-training-data.csv")
titanic.head()
```

```python
titanic.columns =[ "PassengerId","Survived","Pclass","Name","Sex","Age","SibSp","Parch","Ticket","Fare","Cabin","Embarked"]
titanic.columns
```

```python
titanic.isnull().sum()
```

```python
titanic = titanic.drop(['Name','Ticket','Cabin'], axis  = 1)
titanic.head()
```

```python
sns.boxplot(x = 'Parch',y ='Age', data =  titanic)
```

```python
Parch_groups = titanic.groupby(titanic['Parch'])
Parch_groups.mean(numeric_only = True)
```

```python
def age_approx(cols):
    Age = cols.iloc[0]
    Parch = cols.iloc[1]
    
    if pd.isnull(Age):
        if Parch == 0:
            return 32
        elif Parch ==1:
            return 24
        elif Parch == 2:
            return 17
        elif Parch == 3:
            return 33
        elif Parch == 4:
            return 45
        else:
            return 30
        
    else:
        return Age
```

```python
titanic['Age'] = titanic[['Age','Parch']].apply(age_approx,axis = 1)
titanic.isnull().sum()
```

```python
titanic.dropna(inplace = True)
titanic.reset_index(inplace = True,drop = True)

print(titanic.info())
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
gender_cat = titanic['Sex']
gender_encoded = label_encoder.fit_transform(gender_cat)
gender_encoded[0:5]
```

```python
titanic.head()
```

```python
gender_df = pd.DataFrame(gender_encoded,columns = ['male_gender'])
gender_df.head()
```

```python
embarked_cat = titanic['Embarked']
embarked_encoded = label_encoder.fit_transform(embarked_cat)
embarked_encoded[:100]
```

```python
from sklearn.preprocessing import OneHotEncoder
binary_encoder = OneHotEncoder(categories = 'auto')
embarked_1hot = binary_encoder.fit_transform(embarked_encoded.reshape(-1,1))
embarked_1hot_mat = embarked_1hot.toarray()
embarked_df=pd.DataFrame(embarked_1hot_mat,columns = ['C','Q','S'])
embarked_df.head()
```

```python
#titanic.drop(['Sex','Embarked'],axis = 1,inplace = True)
#titanic.head()
```

```python
titanic_dmy = pd.concat([titanic,gender_df,embarked_df],axis = 1,verify_integrity = True).astype(float)
titanic_dmy[0:5]
```

```python
sns.heatmap(titanic_dmy.corr())
```

```python
titanic_dmy.drop(['Fare'],axis = 1,inplace=True)
titanic_dmy.head()
```
